# OpenAI API LLM 

```python
from protolink.llms.api import OpenAILLM
from protolink.models import Message, Part

# Set OPENAI_API_KEY env variable or pass it to the constructor
llm = OpenAILLM(api_key="")
```

```python
# Input

part = Part(type="text", content="Hello, how are you?")
messages = [Message(role="user", parts=[part])]
```

### Test normal response

```python
print("\nTesting non-streaming response:")
response = llm.generate_response(messages)  # message
print(f"Response: {response.to_dict()}")
```

### Test streaming response

```python
print("\nTesting streaming response:")
for chunk in llm.generate_stream_response(messages):
    print(chunk.to_dict().get("parts", ""), end="", flush=True)
    print("\n")
print("\n")
```
